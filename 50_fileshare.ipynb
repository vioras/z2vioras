# ファイル共有

JupyterHubでは、
単一の利用者 `jovyan` だけが利用できる環境にログインすることにより、安全・安心な利用が担保されます。
しかし、システムの制約から共有型ファイル・システムが使えず、また、
Linux環境であるため公認された手段ではOneDriveのような同期型クラウド・ストレージが使えません。
このため、
VIORAS JupyterHubに接続しているPCを介してファイルを手動で共有 (アップロード・ダウンロード) する必要があります。

以下では、VIORAS JupyterHub接続しているPCとファイルを共有する手法を紹介します。
また、プログラミングに際して、非同期型の大規模クラウド・ストレージを直接操作する手法を紹介します。

## JupyterLab Filer Browser

左サイドバーのアイコン列の一番上からファイル・ブラウザを選択します。

```{figure} ./50_fileshare_files/6a178321-dea5-4e90-b08f-403b1bb51530.png
```

### ファイルのアップロード

ファイル・ブラウザのアップロード・アイコンをクリックすると、PCのファイル選択画面が表示されます。
アップロードしたいファイルを選択してください。

```{figure} ./50_fileshare_files/ae3cd649-92ff-4973-b77e-16fe56de2092.png
```

### ファイルのダウンロード

ファイル・ブラウザから、PCにダウンロードしたいファイルを選択します。
このとき右クリックしてコンテキスト・メニューを表示させてください 
(Macでは、コントロールキーを押しながらクリック)。
`Download`を選択すると、PCにファイルがダウンロードされます。

```{figure} ./50_fileshare_files/d735a588-5e2c-4729-b094-5acfeff10f29.png
```

複数のファイルを選択してダウンロードすることもできます。選択方法は、次の回答を参考にして下さい。

### アーカイブ・ファイルのアップロードとダウンロード

VIORAS JupyterHubには機能拡張が導入されているので、ZIPアーカイブされたファイルをアップロードして展開することができます。
また、フォルダのファイルをZIPアーカイブして一括してダウンロードすることができます。

アップロードしたZIPファイルを展開するには、ファイルを右クリックして `Extract Archive` を選択します。

```{figure} ./50_fileshare_files/f189315b-009e-4559-9c32-5e20c7e352a2.png
```

カレント・フォルダのファイルをZIPあーカーブして一括ダウンロードすることができます。
この際、右クリックする場所は、フォルダ内のどこでも (いずれのファイルでも余白でも) 構いません。

```{figure} ./50_fileshare_files/d004e166-2822-43f1-839a-2649fbb71333.png
```
